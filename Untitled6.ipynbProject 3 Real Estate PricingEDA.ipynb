Step 4: Handling Missing Values

üî¢ Step 9: Skewness & Kurtosis Analysis

üìù Explanation for Step 9:

Skewness and Kurtosis Analysis

In this step, we analyzed the **distribution shape** of numerical features using:

- **Skewness**: Measures asymmetry of the distribution.
  - A skewness > 0 indicates right-skewed (positive skew).
  - A skewness < 0 indicates left-skewed (negative skew).
  
- **Kurtosis**: Measures the "tailedness" of the distribution.
  - High kurtosis (> 3) indicates more outliers.
  - Low kurtosis (< 3) indicates fewer outliers than normal distribution.

This analysis helps us decide if data transformation (like log or Box-Cox) is needed to normalize variables for machine learning models.ing models.


‚úÖ Code for Step 9:

```python
from scipy.stats import skew, kurtosis

# Calculate skewness and kurtosis for numerical columns
numerical_cols = df.select_dtypes(include=['number']).columns

for col in numerical_cols:
    sk = skew(df[col].dropna())
    ku = kurtosis(df[col].dropna())
    print(f"{col} -> Skewness: {sk:.2f}, Kurtosis: {ku:.2f}")
```

üî¢ Step 9: Skewness & Kurtosis Analysis

üìù Markdown Explanation for Step 9:

Step 9: Skewness and Kurtosis Analysis

In this step, we assessed the **distribution shape** of each numerical feature using two important statistical metrics:

- **Skewness**: Measures asymmetry in the distribution.
  - A skewness > 0 indicates a **right-skewed** (positively skewed) distribution.
  - A skewness < 0 indicates a **left-skewed** (negatively skewed) distribution.
  - Example: `LotArea` (Skewness = 12.20), `MiscVal` (24.45) are highly skewed.

- **Kurtosis**: Measures how heavy or light the tails are in a distribution.
  - Kurtosis > 3: Heavy tails (more outliers).
  - Kurtosis < 3: Light tails (fewer outliers).
  - Example: `MiscVal` (Kurtosis = 698.60), `PoolArea` (222.50) have extreme kurtosis values.

üëâ This analysis helps detect variables that deviate from normal distribution. It guides us in deciding whether **transformation** (like log, square root, or Box-Cox) is needed before modeling.


‚úÖ Code for Step 9:

```python
from scipy.stats import skew, kurtosis

# Calculate skewness and kurtosis for numerical columns
numerical_cols = df.select_dtypes(include=['number']).columns

for col in numerical_cols:
    sk = skew(df[col].dropna())
    ku = kurtosis(df[col].dropna())
    print(f"{col} -> Skewness: {sk:.2f}, Kurtosis: {ku:.2f}")
```

üì¶üìâ ùêíùêìùêÑùêè ùüèùüé ‚Äì ùêéùêîùêìùêãùêàùêÑùêë ùêÉùêÑùêìùêÑùêÇùêìùêàùêéùêç ùêîùêíùêàùêçùêÜ ùêÅùêéùêóùêèùêãùêéùêìùêí üìâüì¶
